```python
import torch
ssd_model = torch.hub.load('NVIDIA/DeepLearningExamples:torchhub', 'nvidia_ssd')
utils = torch.hub.load('NVIDIA/DeepLearningExamples:torchhub', 'nvidia_ssd_processing_utils')
```

```python
ssd_model.to('cuda')
ssd_model.eval()
```

```python
urls = [
    "/content/04_jpg.rf.c0b83432c4d09c7cbeaac18e14c4a54a.jpg",
    "/content/11_jpg.rf.82fc4a4feafb6d675fb342136f9cf1bf.jpg",
    "/content/12_jpg.rf.ac83fcb8cc8c8bbc5587c2e6881e3d4a.jpg",
    "/content/13_jpg.rf.aa7b6967ac3058c4e3da87e64f9a5143.jpg",
    "/content/14_jpg.rf.3e4fbd917017cceb436ba80692f6512c.jpg"
]
inputs = [utils.prepare_input(uri) for uri in urls]
tensor = utils.prepare_tensor(inputs)
```

```python
with torch.no_grad():
    detections_batch = ssd_model(tensor)


results_per_input = utils.decode_results(detections_batch)
best_results_per_input = [utils.pick_best(results, 0.40) for results in results_per_input]
```

```python
classes_to_labels = utils.get_coco_object_dictionary()
```

```python
from matplotlib import pyplot as plt
import matplotlib.patches as patches

for image_idx in range(len(best_results_per_input)):
    fig, ax = plt.subplots(1)
    # Show original, denormalized image...
    image = inputs[image_idx] / 2 + 0.5
    ax.imshow(image)
    # ...with detections
    bboxes, classes, confidences = best_results_per_input[image_idx]
    for idx in range(len(bboxes)):
        left, bot, right, top = bboxes[idx]
        x, y, w, h = [val * 300 for val in [left, bot, right - left, top - bot]]
        rect = patches.Rectangle((x, y), w, h, linewidth=1, edgecolor='r',\
         facecolor='none')
        ax.add_patch(rect)
        ax.text(x, y, "{} {:.0f}%".format(classes_to_labels[classes[idx] - 1],\
         confidences[idx]*100), bbox=dict(facecolor='white', alpha=0.5))
plt.show()
```
